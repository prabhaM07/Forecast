```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
import numpy as np
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
services = ['Local Route', 'Light Rail', 'Peak Service', 'Rapid Route', 'School']
forecast_days = 7

for service in services:
    df_prophet = df[['Date', service]].rename(columns={'Date': 'ds', service: 'y'})
    df_prophet = df_prophet.dropna()

    model = Prophet(daily_seasonality=True, yearly_seasonality=True, weekly_seasonality=True)
    model.fit(df_prophet)

    future = model.make_future_dataframe(periods=forecast_days)
    forecast = model.predict(future)

    plt.figure(figsize=(10, 4))
    model.plot(forecast)
    plt.title(f'{service} - Prophet Forecast (All Available Data)')
    plt.xlabel('Date')
    plt.ylabel('Number of Journeys')
    plt.tight_layout()
    plt.show()

    forecast_next_7 = forecast[['ds', 'yhat']].tail(forecast_days).reset_index(drop=True)
    forecast_next_7['yhat'] = forecast_next_7['yhat'].apply(lambda x: max(0, x))

    print(f"7-Day Forecast for {service}:")
    print(forecast_next_7.to_string(index=False, header=['Date', 'Forecasted Journeys']))
    print()

    historical_values = df_prophet['y']
    recent_values = historical_values.tail(7)

    recent_avg = recent_values.mean()
    overall_avg = historical_values.mean()

    if recent_avg > overall_avg * 1.05:
        trend = "increasing demand"
    elif recent_avg < overall_avg * 0.95:
        trend = "decreasing demand"
    else:
        trend = "stable demand"

    print(f"Insight for {service}:")
    print(f"- Recent 7-day average journeys: {recent_avg:.0f}")
    print(f"- Overall average journeys: {overall_avg:.0f}")
    print(f"- Forecast indicates {trend}.\n")
```
